```python
import pandas as pd
import numpy as np
from statistics import mean, median, mode ,stdev
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    78,85,90,88,76
}
avg_score = mean(data)
print(avg_score)
mid_score = median(data)
print(mid_score)
mostfreq_score = mode(data)
print(mostfreq_score)


standard_dev = stdev(data)
print(standard_dev)
plt.figure(figsize=(6,4))

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Study_Hours': [2, 3, 5, 7, 8, 4, 9, 10, 1],
    'Exam_Score': [50, 60, 75, 85, 90, 65, 80, 95, 45]
}

df = pd.DataFrame(data)

print("Correlation Coefficient: ", df.corr().iloc[0, 1])


plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['Study_Hours'], y=df['Exam_Score'], color='blue')

plt.title("Study Hours vs. Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")


plt.show()
```

```python
plt.figure(figsize=(4, 4))

sns.boxplot(x=pd.cut(df['Study_Hours'], bins=3, labels=["Low", "Medium", "High"]),
            y=df['Exam_Score'])

plt.title("Study Hours Category vs. Exam Score")
plt.xlabel("Study Hours (Binned)")
plt.ylabel("Exam Score")
plt.show()
```

```python

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')
data = {
    'Study_Hours': [2, 3, 5, 7, 8, 4, 9, 10, 1],
    'Exam_Score': [50, 60, 75, 85, 90, 65, 80, 95, 45],
    'Attendance': [8, 7, 6, 5, 5, 7, 6, 4, 8]
}

df = pd.DataFrame(data)
ax.scatter(df['Study_Hours'], df['Attendance'], df['Exam_Score'], c='blue', marker='o')

ax.set_xlabel('Study Hours')
ax.set_zlabel('Attendance')
ax.set_ylabel('Exam Score')


plt.title("3D Scatter Plot: Study Hours vs Exam Score vs Sleep Hours")

plt.show()
```
